#### Using Deep Learning technique

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
le=le.fit(df['Attrition_Flag'])
df['Attrition_Flag']=le.transform(df['Attrition_Flag'])
```

```python
df=pd.get_dummies(df,columns=['Gender', 'Education_Level', 'Marital_Status', 'Income_Category','Card_Category'])
```

```python
from sklearn.model_selection import train_test_split
x=df.iloc[:,1:]
y=df.iloc[:,0]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,shuffle=True)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train=ss.fit_transform(x_train)
```

```python
from keras.models import Sequential 
from keras.layers import Dense
from keras.layers import Dropout
```

```python
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
def create_model(layers):
    model=Sequential() 
    
    for i ,nodes in enumerate(layers):
        if i==0:
            model.add(Dense(nodes,input_dim=x_train.shape[1],activation='relu',kernel_initializer='uniform'))
        else:
            model.add(Dense(nodes))
        
        model.add(Dropout(0.2))
        model.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))
        
        model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
        
        return model
```

```python
model=KerasClassifier(build_fn=create_model,verbose=0)
```

```python
param_grid=dict(layers=[(37,20)],batch_size=[128],nb_epoch=[50])
grid=GridSearchCV(param_grid=param_grid,cv=5,estimator=model)
result=grid.fit(x_train,y_train)
print(result.best_score_,result.best_params_)
```
